```python
import re
import pandas as pd
```

```python
def import_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text_data = file.read()
        return text_data
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None
    except Exception as e:
        print(f"Error: An error occurred while reading the file - {e}")
        return None

file_path = 'Data/Raja.txt'
text_variable = import_text_file(file_path)

if text_variable is not None:
    print(text_variable)
```

```python
print(type(text_variable))
```

```python
#regex.com
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
#Started from 1 as an empty string is appearing in the start
messages = re.split(pattern, text_variable)[1:]
messages
```

```python
dates = re.findall(pattern, text_variable)
dates
```

```python
# Creating DataFrame
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Convert message date datatype
df['date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ')
df.drop(columns=['message_date'], inplace=True)
```

```python
df.head()
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)  # Added a comma and a '+' after \s
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['messages'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['minute'] = df['date'].dt.minute
```
